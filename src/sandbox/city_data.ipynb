```python
def find_city_site(row):
    city=row["Név"]

    print(city)
    query = city + " online"

    pause=np.random.random()*0.5+0.5
    
    site=next(search(query, tld="co.in", num=1, stop=1, pause=pause))
    row["site"]=site
    return row
```

```python
site_df=df.apply(find_city_site,axis=1)
site_df
```

```python
site_df.to_excel("../data/local/cities_with_site.xlsx")
```
