```python
def close_towns(epicenter, radius):
    for i in town_locations.index:
        row=town_locations.iloc[i]
        coords = (row["Longitude"], row["Latitude"])
        dist = distance.distance(epicenter,coords).km
        
        if dist < radius:
            print (row["Settlement"])

def close_towns_export(epicenter, radius):
    locations=town_locations.groupby("Settlement").mean().reset_index()
    # for i in locations.index:
    #     row=locations.iloc[i]
    #     coords = (row["Longitude"], row["Latitude"])
    #     dist = distance.distance(epicenter,coords).km
        
    #     if dist < radius:
    #         # print (row["Settlement"])
    #         print(row["Settlement"], *coords)
    locations["distance"]=locations.apply(lambda row: distance.distance(epicenter,(row["Longitude"], row["Latitude"])).km,axis=1)
    locations=locations[locations["distance"]<radius]
    return locations
```
