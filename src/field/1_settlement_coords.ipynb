```python
from geopy.geocoders import Nominatim
import pandas as pd
import json
import csv
import tqdm

with open("toosz_towns.json") as f:
    town_urls=json.load(f)


geolocator = Nominatim(user_agent="local_relief")
```

```python
with open("settlement_coordinates.csv","w") as f:
    print ("Settlement,Longitude,Latitude",file=f)
```

```python
with open("settlement_coordinates.csv","a") as f:
    all_towns=list(town_urls.keys())
    for t in tqdm.tqdm(all_towns):
        if t in settlements_done:
            continue
        
        country_town=f"Hungary, {t}"
        geocode=geolocator.geocode(country_town)
        if geocode:
            coords=geocode[1]
            print (f"{t},{coords[0]},{coords[1]}",file=f)
```
