```python
def find_county_site(row):
    county=row["Megye neve"]

    print(county)
    query = county + " online"
    
    site=next(search(query, tld="co.in", num=1, stop=1, pause=2))
    row["site"]=site
    return row
```

```python
site_df=df.apply(find_county_site,axis=1)
site_df
```

```python
site_df.to_excel("../data/local/counties.xlsx")
```

```python
def get_site_name(row):
    row["site_name"]=row["site"].split(".")[-2]

    return row
```

```python
site_df=pd.read_excel("../data/local/counties.xlsx",index_col=0)
site_df=site_df.apply(get_site_name,axis=1)
site_df.head()
```
